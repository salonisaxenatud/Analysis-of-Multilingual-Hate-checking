```python
import pandas as pd
from datasets import Dataset
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
import numpy as np

# Load CSVs
df_hate = pd.read_csv("/content/Hate_pol-eng-hindi.csv")
df_non = pd.read_csv("/content/non-hate-polish-eng-hindi.csv")
df_hate["label"] = 1
df_non["label"] = 0

# Combine and melt all languages into one column
df = pd.concat([df_hate, df_non])
texts = pd.concat([
    df[["Original(Polish)", "label"]].rename(columns={"Original(Polish)": "text"}),
    df[["Translation(English)", "label"]].rename(columns={"Translation(English)": "text"}),
    df[["Translation(Hindi)", "label"]].rename(columns={"Translation(Hindi)": "text"})
]).dropna().reset_index(drop=True)

# Train/test split
train_df = texts.sample(frac=0.6, random_state=42)
val_df = texts.drop(train_df.index)

# Convert to HF Dataset
train_ds = Dataset.from_pandas(train_df.reset_index(drop=True))
val_ds = Dataset.from_pandas(val_df.reset_index(drop=True))

# Load BLOOMZ
model_name = "xlm-roberta-base"
 # or "bigscience/bloomz-7b1"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=2)

# Tokenization
```

```python
def tokenize_fn(examples):
    return tokenizer(examples["text"], truncation=True, padding=True)

train_ds = train_ds.map(tokenize_fn, batched=True)
val_ds = val_ds.map(tokenize_fn, batched=True)

# Metrics
def compute_metrics(eval_pred):
    preds = np.argmax(eval_pred.predictions, axis=1)
    labels = eval_pred.label_ids
    return {
        "accuracy": accuracy_score(labels, preds),
        "f1": f1_score(labels, preds),
        "precision": precision_score(labels, preds),
        "recall": recall_score(labels, preds)
    }

# Training config
```

```python
import pandas as pd
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM
from torch.nn.functional import log_softmax
from tqdm import tqdm

# Load LLaMA model and tokenizer
model_id = "huggyllama/llama-7b" # change if you're using a different variant
tokenizer = AutoTokenizer.from_pretrained(model_id, use_fast=False)
model = AutoModelForCausalLM.from_pretrained(model_id, device_map="auto")
model.eval()

# Load dataset
df = pd.read_csv("/content/Hindi_final_combined.csv")

# Hate speech definition
hate_speech_defs = {
    "group_def": (
        "Definition of hate speech:\n"
        "\"Hate is defined as abusive speech targeting specific group characteristics, "
        "such as ethnic origin, religion, gender, or sexual orientation.\""
    )
}

# Language examples
language_examples = {
    "english": (
        'Example 1: "All Muslims are terrorists." → Yes\n'
        'Example 2: "I don’t like this movie." → No'
    ),
    "polish": (
        'Przykład 1: "Wszyscy Żydzi to złodzieje." → Tak\n'
        'Przykład 2: "Nie lubię tej książki." → Nie'
    ),
    "hindi": (
        'उदाहरण 1: "सभी मुस्लिम आतंकवादी हैं।" → हाँ\n'
        'उदाहरण 2: "मुझे यह गाना पसंद नहीं आया।" → नहीं'
    )
}

# Yes/No tokens per language
yesno_tokens = {
    "english": [" Yes", " No"],
    "polish": [" Tak", " Nie"],
    "hindi": [" हाँ", " नहीं"]
}

# Language prompt suffix
prompt_suffix = {
    "english": '\n\nText: "{text}"\nAnswer with "Yes" or "No":',
    "polish": '\n\nTekst: "{text}"\nOdpowiedz "Tak" lub "Nie":',
    "hindi": '\n\nपाठ: "{text}"\nउत्तर दें: "हाँ" या "नहीं":'
}

# Function to compute logits
def get_yes_no_logits(prompt, lang):
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    with torch.no_grad():
        outputs = model(**inputs)
        logits = outputs.logits[:, -1, :]  # last token prediction
        log_probs = log_softmax(logits, dim=-1)
        yes_id = tokenizer(yesno_tokens[lang][0], add_special_tokens=False)["input_ids"][0]
        no_id = tokenizer(yesno_tokens[lang][1], add_special_tokens=False)["input_ids"][0]
        return log_probs[0, yes_id].item(), log_probs[0, no_id].item()

# Run across languages
for variant_name, definition in hate_speech_defs.items():
    for lang in ["english", "polish", "hindi"]:
        print(f"\n▶️ Processing: {lang.upper()} | Variant: {variant_name}")
        preds, logit_yes, logit_no = [], [], []

        for text in tqdm(df[lang], desc=f"{variant_name} | {lang}"):
            if pd.isna(text) or not str(text).strip():
                preds.append("EMPTY")
                logit_yes.append(None)
                logit_no.append(None)
                continue

            example = language_examples[lang]
            question = prompt_suffix[lang].format(text=text)
            prompt = f"{definition}\n{example}{question}"

            try:
                score_yes, score_no = get_yes_no_logits(prompt, lang)
                prediction = "Yes" if score_yes > score_no else "No"
                preds.append(prediction)
                logit_yes.append(score_yes)
                logit_no.append(score_no)
            except Exception as e:
                preds.append("ERROR")
                logit_yes.append(None)
                logit_no.append(None)
                print(f"[Error] Row failed for {lang} ({variant_name}): {e}")

        df[f"output_{variant_name}_{lang}"] = preds
        df[f"logit_{variant_name}_{lang}_yes"] = logit_yes
        df[f"logit_{variant_name}_{lang}_no"] = logit_no

# Save final results
df.to_csv("llama_yesno_group_def_predictions.csv", index=False)
print("\n✅ Saved to 'llama_yesno_group_def_predictions.csv'")
```
